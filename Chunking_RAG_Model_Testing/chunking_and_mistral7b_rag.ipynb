```python
from langchain.schema import Document

documents = []
for i in data:
    doc = Document(
        page_content=i['text'],
        metadata={"page": int(i['page_num'])}
    )
    documents.append(doc)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=200)
chunks = splitter.split_documents(documents)
```

```python
import re

def extract_section_number(text: str) -> str | None:
    # Match floating numbers like "1.3", "2.5", "3.2.1"
    match = re.search(r'\b\d+(\.\d+)+\b', text)
    return section_metadata.get(match.group(0)) if match else None
'''def extract_section(text):
    match = re.search(r'(Chapter\s?\d+|Section\s?\d+(\.\d+)*)', text, re.IGNORECASE)
    return match.group(0) if match else "Unknown"
'''
for chunk in chunks:
    chunk.metadata["section"] = extract_section_number(chunk.page_content)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA


model_name = "mistralai/Mistral-7B-Instruct-v0.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype="auto")

# Create a pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=512)

llm = HuggingFacePipeline(pipeline=pipe)

# RAG with local model
qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever, return_source_documents=True)

query = "What are the key components of emotion?"
result = qa_chain(query)

print("\n📘 Answer:\n", result['result'])
```
