```python
# Normalize the section titles from section_metadata
section_titles = list(section_metadata.values())
section_titles_clean = [title.lower().strip() for title in section_titles]

current_heading = "Introduction to Psychology"  # default heading (or use section_metadata["1"])

for doc in documents:
    page_text = doc.page_content.lower()

    # Check if any known section title appears in the text
    matched = False
    for title in section_titles_clean:
        if title in page_text:
            current_heading = title  # update heading
            matched = True
            break

    # Assign the current heading to metadata
    doc.metadata["sections"] = current_heading

```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=200)
chunks = splitter.split_documents(documents)
'''for chunk in chunks:
    chunk.metadata["section"] = extract_section_number(chunk.page_content)'''
```

```python
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vectorstore = FAISS.from_documents(chunks, embeddings)

# Save locally
vectorstore.save_local("my_vector_index")
```

```python
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Load the FAISS index
vectorstore = FAISS.load_local(
    "my_vector_index",
    embeddings=embedding_model,
    allow_dangerous_deserialization=True
)
```

#using manual retrieval and retrievor

```python
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 3})
```

```python
query = "What are the key components of emotion?"
docs = retriever.get_relevant_documents(query)

for i, doc in enumerate(docs):
    print(f"\n--- Result {i+1} ---")
    print(f"Page: {doc.metadata.get('page')}")
    print(f"Section: {doc.metadata.get('sections')}")
    print(f"Content: {doc.page_content}")
```
